# Acessando a API

- Necessário cadastro e autenticação

Chave usada (apiKey): 528d65099fb24bf6a49142f9af54ed02

- Há três forma de fazer a autenticação:

    ### apiKey como parâmetro no query da url:

    ```https://newsapi.org/v2/everything?q=keyword&apiKey=528d65099fb24bf6a49142f9af54ed02```

    ### chave X-Api-Key no headers:

    ```{ "X-Api-Key": "528d65099fb24bf6a49142f9af54ed02" }```

    ### chave Authorization no headers:

    ```{ "Authorization": "528d65099fb24bf6a49142f9af54ed02" }```


````python
import requests
import json
import pandas as pd

response = requests.get("https://newsapi.org/v2/everything?q=rna", headers={"X-Api-Key": "528d65099fb24bf6a49142f9af54ed02"})
data = json.loads(response.text)

data["articles"]

df = pd.DataFrame(data["articles"])

display(df)
````
